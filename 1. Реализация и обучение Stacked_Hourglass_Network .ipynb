```python
import albumentations as A

# напишем датасет, который возвращает картинку и хитмапы ключевых точек
class CelebADataset(Dataset):
    def __init__(self, images_dir, landmarks_data, img_size, transform=None, crop_margin=0.2, mode='train'):
        self.images_dir = images_dir
        self.landmarks_data = landmarks_data
        self.img_size = img_size
        self.transform = transform
        self.crop_margin = crop_margin
        self.mode = mode

    def __len__(self):
        return len(self.landmarks_data)

    def __getitem__(self, idx):
        img_name = self.landmarks_data.iloc[idx, 0]
        img_path = os.path.join(self.images_dir,img_name)
        image = cv2.imread(img_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # получаем ключевые точки, обрезаем картинку вокруг них, вычисляем ключевые точки на обрезанной картинке
        keypoints = self.landmarks_data.iloc[idx, 1:].values.astype('float').reshape(-1, 2)
        image, keypoints = self.crop_and_resize(image, keypoints)

        image = image / 255.0

        train_augmentations = A.Compose(
          [
              A.HorizontalFlip(p=0.5),
              A.ShiftScaleRotate(shift_limit=0.0625, scale_limit=0.1, rotate_limit=15, p=0.5),
              A.RandomBrightnessContrast(p=0.2),
          ],
          keypoint_params=A.KeypointParams(format='xy')
        )

        if (self.mode == 'train'):
          augmented = train_augmentations(image=image, keypoints=keypoints)
          image = augmented['image']
          keypoints = augmented['keypoints']

        heatmaps = self.generate_heatmaps(keypoints, sigma=5)

        # тут 3-м и 4-м я возвращаю фактические ключевые точки и имя файла картинки
        # ключевые точки нужны для показа их на картинке
        # имя файла понадобится в конце для сохранения выровненных картинок
        return torch.tensor(image, dtype=torch.float32).permute(2, 0, 1), torch.tensor(heatmaps, dtype=torch.float32), keypoints, img_name

    # метод обрезает картинку вокруг ключевых точек, приводит к нужному размеру и возвращает преобразованную картинку и соотвествующие ей ключевые точки
    def crop_and_resize(self, image, keypoints):
        h, w, _ = image.shape

        x_min = max(0, int(np.min(keypoints[:, 0]) - self.crop_margin * w))
        y_min = max(0, int(np.min(keypoints[:, 1]) - self.crop_margin * h))
        x_max = min(w, int(np.max(keypoints[:, 0]) + self.crop_margin * w))
        y_max = min(h, int(np.max(keypoints[:, 1]) + self.crop_margin * h))

        cropped_image = image[y_min:y_max, x_min:x_max]

        keypoints[:, 0] -= x_min
        keypoints[:, 1] -= y_min

        scale_x = self.img_size / (x_max - x_min)
        scale_y = self.img_size / (y_max - y_min)
        resized_image = cv2.resize(cropped_image, (self.img_size, self.img_size))
        keypoints[:, 0] *= scale_x
        keypoints[:, 1] *= scale_y

        return resized_image, keypoints

    # метод для генерации хитмапов по ключевым точкам
    def generate_heatmaps(self, keypoints, sigma):
      """Generate heatmaps for each keypoint."""
      heatmaps = np.zeros((len(keypoints), img_size, img_size))
      for i, (x, y) in enumerate(keypoints):
          if x < 0 or y < 0 or x >= self.img_size or y >= self.img_size:
              continue

          xx, yy = np.meshgrid(np.arange(self.img_size), np.arange(self.img_size))
          heatmap = np.exp(-((xx - x)**2 + (yy - y)**2) / (2 * sigma**2))
          heatmaps[i] = heatmap
      return heatmaps
```

```python
# реализуем StackedHourglassNetwork
class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.skip = nn.Identity() if in_channels == out_channels else nn.Conv2d(in_channels, out_channels, 1)
        mid_channels = out_channels // 2

        self.conv1 = nn.Conv2d(in_channels, mid_channels, 1)
        self.bn1 = nn.BatchNorm2d(mid_channels)
        self.conv2 = nn.Conv2d(mid_channels, mid_channels, 3, padding=1)
        self.bn2 = nn.BatchNorm2d(mid_channels)
        self.conv3 = nn.Conv2d(mid_channels, out_channels, 1)
        self.bn3 = nn.BatchNorm2d(out_channels)

    def forward(self, x):
        residual = self.skip(x)
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.relu(self.bn2(self.conv2(x)))
        x = self.bn3(self.conv3(x))
        x += residual

        return F.relu(x)


class Hourglass(nn.Module):
    def __init__(self, in_channels, depth):
        super(Hourglass, self).__init__()
        self.depth = depth
        self.downsample = nn.Conv2d(in_channels, in_channels, kernel_size=3, stride=2, padding=1)

        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=False)

        self.down_blocks = nn.ModuleList([ResidualBlock(in_channels, in_channels) for _ in range(depth)])
        self.up_blocks = nn.ModuleList([ResidualBlock(in_channels, in_channels) for _ in range(depth)])
        self.bottom_block = ResidualBlock(in_channels, in_channels)

    def forward(self, x):
        skip_connections = []

        for down_block in self.down_blocks:
            x = down_block(x)
            skip_connections.append(x)
            x = self.downsample(x)

        x = self.bottom_block(x)

        for up_block in self.up_blocks:
            x = self.upsample(x)
            x = x + skip_connections.pop()
            x = up_block(x)

        return x

class StackedHourglassNetwork(nn.Module):
    def __init__(self, in_channels, num_keypoints, num_stacks=4, depth=3):
        super(StackedHourglassNetwork, self).__init__()
        self.num_stacks = num_stacks

        self.preprocess = nn.Sequential(
            nn.Conv2d(in_channels, 128, kernel_size=7, stride=1, padding=3),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            ResidualBlock(128, 128)
        )

        self.hourglasses = nn.ModuleList([
            Hourglass(128, depth=depth) for _ in range(num_stacks)
        ])

        self.intermediate_convs = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(128, 128, kernel_size=1),
            ) for _ in range(num_stacks - 1)
        ])

        self.output_convs = nn.ModuleList([
            nn.Conv2d(128, num_keypoints, kernel_size=1) for _ in range(num_stacks)
        ])

        self.skip_convs = nn.ModuleList([
            nn.Conv2d(num_keypoints, 128, kernel_size=1, bias=False)
            for _ in range(num_stacks - 1)
        ])

    def forward(self, x):
        x = self.preprocess(x)
        outputs = []

        for i in range(self.num_stacks):
            x = self.hourglasses[i](x)

            output = self.output_convs[i](x)
            outputs.append(output)

            if i < self.num_stacks - 1:
                inter = self.intermediate_convs[i](x)
                skip = self.skip_convs[i](output)
                x = x + inter + skip

        return outputs
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
from torch.optim.lr_scheduler import ReduceLROnPlateau

model = StackedHourglassNetwork(in_channels=3, num_keypoints=5, num_stacks=3).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.MSELoss()
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=10)
```

```python
# функция для рисования лоса
def plot_loss_history(loss_history):
  plt.plot(range(1, len(loss_history) + 1), loss_history, label='Loss')
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.title('Training Loss History')
  plt.legend()
  plt.show()

# функция для преобразования хитмапов в ключевые точки
def heatmaps_to_keypoints(heatmaps):
    keypoints = []
    for heatmap in heatmaps:
        y, x = np.unravel_index(np.argmax(heatmap), heatmap.shape)
        keypoints.append((x, y))
    return np.array(keypoints)
```

```python
# реализуем обучение с валидацией после каждой эпохи с сохранением лучшей модели
# для наглядности будем рисовать картинку с предсказанными ключевыми точками каждые 5 эпох
train_loss_history = []
val_loss_history = []
best_val_loss = float('inf')

for epoch in range(60):
    model.train()
    train_loss = 0.0
    for images, heatmaps, _, _ in train_data_loader:
      images, heatmaps = images.to(device), heatmaps.to(device)
      optimizer.zero_grad()
      outputs = model(images)
      loss = sum(criterion(output, heatmaps) for output in outputs)
      loss.backward()
      optimizer.step()

      train_loss += loss.item()

    avg_train_loss = train_loss / len(train_data_loader)
    print(f"Epoch {epoch + 1}, train loss: {avg_train_loss}")
    train_loss_history.append(avg_train_loss)

    model.eval()
    val_loss = 0.0
    with torch.no_grad():
      for images, heatmaps, _, _ in val_data_loader:
        images, heatmaps = images.to(device), heatmaps.to(device)
        outputs = model(images)
        loss = sum(criterion(output, heatmaps) for output in outputs)
        val_loss += loss.item()
      if (epoch + 1) % 5 == 0:
        pred_heatmaps = outputs[-1].cpu()
        pred_heatmap_for_first_image = pred_heatmaps[0].cpu().numpy()
        pred_keypoints_for_first_image = heatmaps_to_keypoints(pred_heatmap_for_first_image)
        first_img = images[0].cpu().permute(1, 2, 0).numpy()
        plot_image_with_keypoints(first_img, pred_keypoints_for_first_image)

    avg_val_loss = val_loss / len(val_data_loader)

    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        torch.save(model.state_dict(), "best_stacked_hourglass_model.pth")
        print(f"Model saved at epoch {epoch+1} with val loss {best_val_loss:.4f}")
    val_loss_history.append(avg_val_loss)

    scheduler.step(avg_val_loss)
```

```python
# посмотрим, как лучшая модель научилась предсказывать ключевые точки
my_model = StackedHourglassNetwork(in_channels=3, num_keypoints=5, num_stacks=3).to(device)
my_model.load_state_dict(torch.load("best_stacked_hourglass_model.pth"))

my_model.eval()
with torch.no_grad():
  for images, _, _, _ in val_data_loader:
    images = images.to(device)
    outputs = my_model(images)
    heatmaps = outputs[-1].cpu()
    for i in range(batch_size):
      heatmap = heatmaps[i].cpu().numpy()
      pred_keypoints = heatmaps_to_keypoints(heatmap)
      img = images[i].cpu().permute(1, 2, 0).numpy()
      plot_image_with_keypoints(img, pred_keypoints)
    break;
```

```python
# этот даталоадер мне нужен для работы с данными из трейн части, но без аугментаций
train_dataset_no_augmentations = CelebADataset(images_dir, train_data, img_size=img_size, crop_margin=0.2, mode='test')
train_data_loader_no_augmentations = DataLoader(train_dataset_no_augmentations, batch_size=batch_size, shuffle=True, num_workers=2)
```

```python
# для выравнивания посчитаю средние значения по всему датасету для фактических координат ключевых точек, чтобы вычислить target_points
target_points = torch.zeros(5, 2)
loaders = [train_data_loader_no_augmentations, val_data_loader]
with torch.no_grad():
  for loader in loaders:
    for _, _, keypoints,_ in loader:
      target_points += torch.sum(keypoints, 0)

target_points = target_points / 6038
target_points = torch.round(target_points)
target_points = target_points.numpy()
target_points
```

```python
# посмотрим насколько "на месте" выглядят target_points на картинках
my_model.eval()
with torch.no_grad():
  for images, _, _ , _ in train_data_loader_no_augmentations:
    images = images.to(device)
    for i in range(batch_size):
      img = images[i].cpu().permute(1, 2, 0).numpy()
      plot_image_with_keypoints(img, target_points)
    break;
```

```python
# функция для выравнивания картинки по ключевым точкам. В данном случае я беру 3 (глаза и нос)
def align_face(image, keypoints):
  affine_matrix, _ = cv2.estimateAffinePartial2D(keypoints[0:3], target_points[0:3])
  aligned_image = cv2.warpAffine(img, affine_matrix, (64, 64))

  return aligned_image

# посмотрим, как выглядят картинки до и после выравнивания
my_model.eval()
with torch.no_grad():
  for images, _, _ , _ in val_data_loader:
    images = images.to(device)
    outputs = my_model(images)
    heatmaps = outputs[-1].cpu()
    for i in range(batch_size):
      heatmap = heatmaps[i].cpu().numpy()
      pred_keypoints = heatmaps_to_keypoints(heatmap)
      img = images[i].cpu().permute(1, 2, 0).numpy()
      aligned_image = align_face(img, pred_keypoints)
      fig, axes = plt.subplots(1, 2, figsize=(4*2, 4*2))
      axes[0].imshow(img);
      axes[0].set_title("Original Image")

      axes[1].imshow(aligned_image);
      axes[1].set_title("Aligned Image")
    break;
```

```python
# создадим папку для выровненных картинок
output_dir = "aligned_images"
os.makedirs(output_dir, exist_ok=True)
```

```python
# переместим в нее все картинки из val_data_loader и train_data_loader
my_model.eval()
loaders = [train_data_loader_no_augmentations, val_data_loader]
with torch.no_grad():
  for loader in loaders:
    for batch_idx, (images, _, _, original_names) in enumerate(loader):
      images = images.to(device)
      outputs = my_model(images)
      heatmaps = outputs[-1].cpu()
      for i in range(len(images)):
        heatmap = heatmaps[i].cpu().numpy()
        pred_keypoints = heatmaps_to_keypoints(heatmap)
        img = (images[i].cpu().permute(1, 2, 0).numpy() * 255).astype(np.uint8)
        original_name = original_names[i]
        aligned_image = align_face(img, pred_keypoints)
        output_path = os.path.join(output_dir, original_name)
        cv2.imwrite(output_path, cv2.cvtColor(aligned_image, cv2.COLOR_RGB2BGR))
```
