```python
import pandas as pd
import os
import shutil
```

```python
# для решения задачи классификации нам нужно отобрать такие классы, у которых число картинок будет максимальным
identity_csv = "identity_CelebA.csv"
identity_df = pd.read_csv(identity_csv)
# возьмем первые 200 самых встречающихся классов
top_classes = identity_df['id'].value_counts().head(200).index
# отфильтруем в identity_df только те данные, которые соответствуют top_classes
filtered_identity_df = identity_df[identity_df['id'].isin(top_classes)]
filtered_identity_df
```

```python
# сохраним как identity_CelebA_mini.csv
filtered_identity_df.to_csv('identity_CelebA_mini.csv', index=False)
```

```python
# аналогичным образом отфильтруем данные по ключевым точкам
landmarks_csv_path = 'list_landmarks_align_celeba.csv'
landmarks_data = pd.read_csv(landmarks_csv_path)
landmarks_data_filtered = landmarks_data[landmarks_data['image_id'].isin(filtered_identity_df['image_id'])]
landmarks_data_filtered
```

```python
# сохраним
landmarks_data_filtered.to_csv('list_landmarks_align_celeba_mini.csv', index=False)
```

```python
# функция для переноса файлов из одной папки в другую, используя столбец image_id во фрейме df
def move_files(df, source_dir, dest_dir):
    for _, row in df.iterrows():
        src = os.path.join(source_dir, row['image_id'])
        dest = os.path.join(dest_dir, row['image_id'])
        shutil.copy(src, dest)
        
# img_align_celeba.zip можно скачать c https://drive.google.com/drive/folders/0B7EVK8r0v71pTUZsaXdaSnZBZzg?resourcekey=0-rJlzl934LzC-Xp28GeIBzQ,
# но я не стал его добавлять в гитхаб, так как много весит
# создадим папку 'img_align_celeba_mini' для переноса картинок, соотвествующих filtered_identity_df, и скопируем их туда из img_align_celeba
# !unzip -q img_align_celeba.zip
os.makedirs('img_align_celeba_mini', exist_ok=True)
move_files(filtered_identity_df, 'img_align_celeba', 'img_align_celeba_mini') 
```
